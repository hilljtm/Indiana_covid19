```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 7))

# Load the data
df = pd.read_csv("Data/covid19_county.csv")

# Plot the total cases
sns.set_color_codes("pastel")
sns.barplot(x=df["Cases_17"], y=df["County"], data=df,
            label="Total", color="b")
sns.barplot(x=df["Cases_18"], y=df["County"], data=df,
            label="Total", color="r")
sns.barplot(x=df["Cases_19"], y=df["County"], data=df,
            label="Total", color="g")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",
       xlabel="Confirmed cases in Indiana by county")
sns.despine(left=True, bottom=True)
```

```python
# Setting the positions and width for the bars

df = pd.read_csv('Data/covid19_county.csv')

pos = list(range(len(df['County']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(15,10))

# Create a bar with Cases_17 data,
# in position pos,
plt.bar(pos, 
        #using df['Cases_17'] data,
        df['Cases_17'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in County
        label=df['County'][0]) 

# Create a bar with Cases_18 data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['Cases_18'] data,
        df['Cases_18'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in County
        label=df['County'][1]) 

# Create a bar with Cases_19 data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['Cases_19'] data,
        df['Cases_19'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in County
        label=df['County'][2]) 

# Set the y axis label
ax.set_ylabel('Cases')

# Set the chart's title
ax.set_title('Covid19 cases in Indiana')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])
plt.xticks(rotation=80)

# Set the labels for the x ticks
ax.set_xticklabels(df['County'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(df['Cases_17'] + df['Cases_18'] + df['Cases_19'])] )

# Adding the legend and showing the plot
plt.legend(['March 17th', 'March 18th', 'March 19th'], loc='upper left')
plt.grid(True)
plt.show()
plt.savefig('covid.png', orientation='landscape', dpi=80)
```
