```python
# Setting the positions and width for the bars

df = pd.read_csv('Data/covid19_county.csv')

pos = list(range(len(df['County']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(15,10))

# Create a bar with Cases_17 data,
# in position pos,
plt.bar(pos, 
        #using df['Cases_17'] data,
        df['Cases_17'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in County
        label=df['County'][0]) 

# Create a bar with Cases_18 data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['Cases_18'] data,
        df['Cases_18'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in County
        label=df['County'][1]) 

# Create a bar with Cases_19 data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['Cases_19'] data,
        df['Cases_19'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in County
        label=df['County'][2]) 

# Create a bar with Cases_20 data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['Cases_20'] data,
        df['Cases_20'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FF5733', 
        # with label the fourth value in County
        label=df['County'][3])

# Create a bar with Cases_21 data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['Cases_21'] data,
        df['Cases_21'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FF5733', 
        # with label the fifth value in County
        label=df['County'][4])



# Set the y axis label
ax.set_ylabel('Cases')

# Set the chart's title
ax.set_title('Covid19 cases in Indiana')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])
plt.xticks(rotation=90)

# Set the labels for the x ticks
ax.set_xticklabels(df['County'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(df['Cases_17'] + df['Cases_18'] + df['Cases_19'] + df['Cases_20'] + df['Cases_21'])] )

# Adding the legend and showing the plot
plt.legend(['March 17th', 'March 18th', 'March 19th', 'March 20th', 'March 21st'], loc='upper left')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import altair as alt

df = pd.read_csv('Data/covid19_county.csv')

source = df

bars = alt.Chart(source).mark_bar(size=5).encode(x='Cases_31', y='County')

text = bars.mark_text(align='left', baseline='middle', dx=3).encode(text='Cases_31').interactive()

(bars + text).properties(height=800, width=700)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")

deaths = pd.read_csv('Data/covid_deaths.csv')

# Set up the matplotlib figure
f, ax1 = plt.subplots(figsize=(15, 10), sharex=True)

# Generate some sequential data
x = deaths['County']
y1 = deaths['Deaths']
sns.barplot(x=x, y=y1, palette="rocket", ax=ax1)
ax1.axhline(0, animated=True,  color="k", clip_on=True)
ax1.set_ylabel("Count")
plt.xticks(rotation=90, horizontalalignment='right')
plt.show();
```
